# **ETL/ELT con Apache Airflow**

## **PROPÓSITO DEL PROYECTO**
El objetivo de este proyecto es desarrollar un pipeline de datos que pueda desarrollar las siguientes tareas:
1.	Consultar la API de Mercado Libre para obtener información relevante sobre productos, como ID, site_id, título, precio y cantidad vendida. Para este proyecto el producto a extraer fueron los microondas.
2.	Cargar los datos obtenidos de la API en una base de datos PostgreSQL
3.	Realizar un análisis de ventas, para este proyecto se optó por verificar si hay productos cuyo precio y cantidad vendida resulten en ganancias superiores a un umbral, como $7000000 por ejemplo.
4.	Envía alertas por correo electrónico si se detectan productos que cumplen con los criterios establecidos en el análisis de ventas.


## **DETALLES DEL PROYECTO**

### **TECNOLOGÍAS UTILIZADAS**


*   Apache Airflow: Utilizado como el motor de ejecución de tareas y programación de trabajos para automatizar flujos de trabajo.

*   Docker: Empleado para crear contenedores de aplicaciones y orquestar el entorno de ejecución de Apache Airflow de manera eficiente y portátil.

*   Visual Studio Code (VSCode): Utilizado como entorno de desarrollo para escribir, depurar y administrar código.
*   Postman: Herramienta utilizada para probar y depurar APIs.


*   PostgreSQL: Base de datos utilizada para almacenar y gestionar los datos adquiridos de la API de Mercado Libre.


*   DBeaver: Herramienta de administración de bases de datos de código abierto que se puede utilizar para conectarse a bases de datos PostgreSQL.



### **DESCRIPCIÓN BREVE DE LOS SCRIPTS UTILIZADOS**


*   **tecnica_postgres.py:** Define y configura el flujo de trabajo con Apache Airflow. Este script orquesta el proceso completo y consta de cuatro tareas secuenciales: crear una tabla en PostgreSQL, consultar la API, insertar datos en la tabla y leer datos para el análisis de ventas.

*   **consult_api.py:** Consulta la API de Mercado Libre para obtener información actualizada sobre productos, como ID, site_id, título, precio y cantidad vendida. Los datos se guardan en un archivo TSV para su posterior carga en la base de datos.
*   **PostgresFileOperator.py:** Se comporta como un operador de Airflow y se utiliza para realizar operaciones de escritura y lectura en la base de datos PostgreSQL. Además, toma decisiones sobre si enviar correos electrónicos de alerta en función de los resultados del análisis de ventas.
